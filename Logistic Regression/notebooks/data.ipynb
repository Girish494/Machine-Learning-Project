```python
import pandas as pd
```

```python
df=pd.read_csv(f"C:\PrimeBatch\AlgoPractice\Logistic Regression\data\\raw_data\data.csv")
df.isnull().value_counts()
```

```python
#Perform EDA 
# The dataset contains NaN values in the column Unnamed: 32, which is not relevant for the model, so we drop this column.
df=df.drop(["Unnamed: 32"],axis=1)
```

```python
#Perform Feature Enginnering
# The dataset contains M and B values in the column diagnosis, which is relevant for the model, M (malignant) indicates person is Cancerous and B  (benign) indicates peson is Non-Cancerous
df["diagnosis"]=df["diagnosis"].map({"M":1,"B":0})
```

```python
df.to_csv("processed_data.csv",index=False)
```
